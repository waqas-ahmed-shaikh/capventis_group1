```python
labels = ['low', 'medium', 'high']

# Create a new column with binned values using qcut to create equal-sized bins
df4['income_category'] = pd.qcut(df4['sales'], q=3, labels=labels)

# Display the DataFrame
df4.head()
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from scipy.stats import chi2_contingency, ttest_ind
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk

# Define sales threshold to categorize high and low sales
sales_threshold = np.median(df4['sales'])

# Categorize data into high sales and low sales
df4['sales_category'] = np.where(df4['sales'] >= sales_threshold, 'high', 'low')

# Function to clean and tokenize text
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Function to clean, tokenize, and lemmatize text
def preprocess_text(text):
    # Remove non-alphanumeric characters
    text = re.sub(r'\W+', ' ', text)
    # Convert text to lowercase
    text = text.lower()
    # Tokenize and remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatize tokens
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    # Join tokens back into a single string
    return ' '.join(tokens)


# Preprocess the 'Introduction' column
df4['cleaned_intro'] = df4['Introduction'].apply(preprocess_text)

# Split data into high sales and low sales groups
high_sales_intros = df4[df4['sales_category'] == 'high']['cleaned_intro']
low_sales_intros = df4[df4['sales_category'] == 'low']['cleaned_intro']

# Vectorize the text data
vectorizer = CountVectorizer()
X_high = vectorizer.fit_transform(high_sales_intros)
X_low = vectorizer.transform(low_sales_intros)

# Get word frequencies
word_freq_high = np.sum(X_high.toarray(), axis=0)
word_freq_low = np.sum(X_low.toarray(), axis=0)
words = vectorizer.get_feature_names_out()

# Create a DataFrame with word frequencies
word_freq_df = pd.DataFrame({
    'word': words,
    'high_sales_freq': word_freq_high,
    'low_sales_freq': word_freq_low
})

# Perform chi-square test to find significant differences
word_freq_df['total'] = word_freq_df['high_sales_freq'] + word_freq_df['low_sales_freq']
word_freq_df['expected_high'] = word_freq_df['total'] * len(high_sales_intros) / len(df)
word_freq_df['expected_low'] = word_freq_df['total'] * len(low_sales_intros) / len(df)

# Chi-square statistic calculation
chi2, p = chi2_contingency(
    [word_freq_df['high_sales_freq'], word_freq_df['low_sales_freq']]
)[:2]

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Check if p-value is less than 0.05 (common threshold for significance)
if p < 0.05:
    print("There is a significant difference between the introductions of high sales and low sales accounts.")
else:
    print("There is no significant difference between the introductions of high sales and low sales accounts.")
```

```python
word_freq_df['chi2_contrib'] = (word_freq_df['high_sales_freq'] - word_freq_df['low_sales_freq'])**2 / (word_freq_df['high_sales_freq'] + word_freq_df['low_sales_freq'])
key_words = word_freq_df.sort_values(by='chi2_contrib', ascending=False).head(10)
print("Key words contributing to the difference:")
key_words.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_high = tfidf_vectorizer.fit_transform(high_sales_intros)
tfidf_low = tfidf_vectorizer.transform(low_sales_intros)

# Get feature names
tfidf_features = tfidf_vectorizer.get_feature_names_out()

# Sum TF-IDF scores for each word
tfidf_sum_high = np.sum(tfidf_high.toarray(), axis=0)
tfidf_sum_low = np.sum(tfidf_low.toarray(), axis=0)

# Create DataFrame
tfidf_df = pd.DataFrame({
    'word': tfidf_features,
    'high_tfidf': tfidf_sum_high,
    'low_tfidf': tfidf_sum_low
})

# Find top TF-IDF scores
top_tfidf_high = tfidf_df.sort_values(by='high_tfidf', ascending=False).head(10)
top_tfidf_low = tfidf_df.sort_values(by='low_tfidf', ascending=False).head(10)

print("Top TF-IDF words in high sales introductions:")
top_tfidf_high.head()

```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Apply sentiment analysis
df4['sentiment'] = df4['cleaned_intro'].apply(get_sentiment)

# Compare sentiment between high and low sales groups
high_sales_sentiment = df4[df4['sales_category'] == 'high']['sentiment']
low_sales_sentiment = df4[df4['sales_category'] == 'low']['sentiment']

print(f"Average sentiment for high sales: {high_sales_sentiment.mean()}")
print(f"Average sentiment for low sales: {low_sales_sentiment.mean()}")

# Statistical test for sentiment difference
t_stat, p_value = ttest_ind(high_sales_sentiment, low_sales_sentiment)

print(f"T-test statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant difference in sentiment between high and low sales introductions.")
else:
    print("There is no significant difference in sentiment between high and low sales introductions.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Use bigrams and trigrams
vectorizer = CountVectorizer(ngram_range=(2, 3))
X_high_ngrams = vectorizer.fit_transform(high_sales_intros)
X_low_ngrams = vectorizer.transform(low_sales_intros)

# Get n-gram frequencies
ngram_freq_high = np.sum(X_high_ngrams.toarray(), axis=0)
ngram_freq_low = np.sum(X_low_ngrams.toarray(), axis=0)
ngrams = vectorizer.get_feature_names_out()

# Create a DataFrame with n-gram frequencies
ngram_freq_df = pd.DataFrame({
    'ngram': ngrams,
    'high_sales_freq': ngram_freq_high,
    'low_sales_freq': ngram_freq_low
})

# Identify key n-grams contributing to the difference
ngram_freq_df['chi2_contrib'] = (ngram_freq_df['high_sales_freq'] - ngram_freq_df['low_sales_freq'])**2 / (ngram_freq_df['high_sales_freq'] + ngram_freq_df['low_sales_freq'])
key_ngrams = ngram_freq_df.sort_values(by='chi2_contrib', ascending=False).head(10)
print("Key n-grams contributing to the difference:")
print(key_ngrams)
```

```python
# Calculate text lengths
df4['intro_length'] = df4['cleaned_intro'].apply(len)

# Compare lengths between high and low sales groups
high_sales_length = df4[df4['sales_category'] == 'high']['intro_length']
low_sales_length = df4[df4['sales_category'] == 'low']['intro_length']

print(f"Average length for high sales introductions: {high_sales_length.mean()}")
print(f"Average length for low sales introductions: {low_sales_length.mean()}")

# Statistical test for length difference
t_stat_length, p_value_length = ttest_ind(high_sales_length, low_sales_length)

print(f"T-test statistic for length: {t_stat_length}")
print(f"P-value for length: {p_value_length}")

if p_value_length < 0.05:
    print("There is a significant difference in the length of high and low sales introductions.")
else:
    print("There is no significant difference in the length of high and low sales introductions.")
```

```python
import textstat

# Function to get readability score
def get_readability(text):
    return textstat.flesch_reading_ease(text)

# Apply readability analysis
df4['readability'] = df4['cleaned_intro'].apply(get_readability)

# Compare readability between high and low sales groups
high_sales_readability = df4[df4['sales_category'] == 'high']['readability']
low_sales_readability = df4[df4['sales_category'] == 'low']['readability']

print(f"Average readability for high sales: {high_sales_readability.mean()}")
print(f"Average readability for low sales: {low_sales_readability.mean()}")

# Statistical test for readability difference
t_stat_readability, p_value_readability = ttest_ind(high_sales_readability, low_sales_readability)

print(f"T-test statistic for readability: {t_stat_readability}")
print(f"P-value for readability: {p_value_readability}")

if p_value_readability < 0.05:
    print("There is a significant difference in readability between high and low sales introductions.")
else:
    print("There is no significant difference in readability between high and low sales introductions.")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Prepare data for clustering
clustering_data = df4[['intro_length', 'readability']]
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(clustering_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Fit KMeans clustering with the optimal number of clusters (let's assume it's 3)
kmeans = KMeans(n_clusters=3, random_state=42)
df4['cluster'] = kmeans.fit_predict(clustering_data_scaled)

# Analyze clusters
sns.pairplot(df4, hue='cluster', vars=['intro_length', 'readability', 'sales'])
plt.show()
```
